```python
class Observable:
    def register_observer(self, observer):
        raise NotImplementedError
    
    def remove_observer(self, observer):
        raise NotImplementedError
    
    def notify_observers(self):
        raise NotImplementedError

class  Observer:
    def update(self, temperature, humidity):
        raise NotImplementedError
        
class DisplayElement:
    def display(self):
        raise NotImplementedError
        
class WeatherData(Observable):
    def __init__(self):
        self._temperature = None
        self._humidity = None
        self._observers = []
        
    def register_observer(self, observer: Observer):
        self._observers.append(observer)
    
    def remove_observer(self, observer: Observer):
        self._observers.remove(observer)
        
    def notify_observers(self):
        for observer in self._observers:
            observer.update(self._temperature, self._humidity)
            
    def set_measurements(self, temperature, humidity):
        self._temperature = temperature
        self._humidity = humidity
        self.notify_observers()

class MobileDisplay(DisplayElement, Observer):
    def __init__(self, weather_data: WeatherData):
        self._temperature = None
        self._humitidy = None
        self._weather_data = weather_data
        weather_data.register_observer(self)
        
    def display(self):
        print(f"Mobile device. The temperature is {self._temperature} and the humidity is {self._humidity}")
            
    def update(self, temp, humidity):
        self._temperature = temp
        self._humidity = humidity
        self.display()
              
class WindowDisplay(DisplayElement, Observer):
    def __init__(self, weather_data: WeatherData):
        self._temperature = None
        self._humitidy = None
        self._weather_data = weather_data
        weather_data.register_observer(self)
    
    def display(self):
        print(f"Window device. The temperature is {self._temperature} and the humidity is {self._humidity}")
            
    def update(self, temp, humidity):
        self._temperature = temp
        self._humidity = humidity
        self.display()

def weather_station():
      weather_data = WeatherData()
      
      mobile_display = MobileDisplay(weather_data)
      window_display = WindowDisplay(weather_data)
      
      weather_data.set_measurements(100, 60)
      weather_data.set_measurements(90, 65)
        
      weather_data.remove_observer(mobile_display)
      weather_data.set_measurements(99,55)

if __name__ == "__main__":
      weather_station()
```
